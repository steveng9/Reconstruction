### QID 1

```python
# Create an MST synthesizer w/ epsilon 1
mst1_ep1_qid1_synthesizer = Synthesizer.create("mst", epsilon=1)
```

```python
# Train the MST synthesizer on the first dataset
mst1_ep1_qid1_synthesizer.fit(
    mst_gt_data_qid_1,
    preprocessor_eps=0.3,
    categorical_columns=relevant_cols["categorical"],
    continuous_columns=relevant_cols["continuous"],
    ordinal_columns=relevant_cols["ordinal"],
)
```

```python
# Create Sample
strata_5_sample = mst1_ep1_qid1_synthesizer.sample(DATA_SIZE)
# Set their data types to int64
for col in strata_5_sample.columns:
    strata_5_sample[col] = (
        strata_5_sample[col].astype(np.int64)
        if strata_5_sample[col].dtype == np.float64
        else strata_5_sample[col]
    )
strata_5_sample
```
